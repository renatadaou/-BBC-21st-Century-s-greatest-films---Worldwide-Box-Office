## Part 1 - testing movies

```python
import re
import requests
from bs4 import BeautifulSoup

head = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
movie = "Mulholland Drive"
all_movie_info = []

clean_movie = re.sub(r"[^\w\d\s]", "", movie)
url = "https://www.boxofficemojo.com/search/?q=" + clean_movie.replace(" ", "+")
raw_html = requests.get(url, headers=head).content
soup_doc = BeautifulSoup(raw_html, "html.parser")
link = soup_doc.find(class_="a-color-secondary").previous_sibling['href']

movie_link = {"name": movie, "link": link}
print(movie_link)
```

### Part 2 - testing more than one movie

```python
head={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
import time
all_movie_info = []
movies = df.movie.unique()[:10]
for movie in movies[:10]:
#modified in loop
    clean_movie = re.sub(r"[^\w\d\s]","",movie)
    url ="https://www.boxofficemojo.com/search/?q=" + clean_movie.replace(" ","+") 
    print(url)
    raw_html = requests.get(url,headers=head).content
    time.sleep(2)
    soup_doc = BeautifulSoup(raw_html, "html.parser")
    try:
        link = soup_doc.find(class_="a-color-secondary").previous_sibling['href']
        link_dict = {"movie": movie, "link": link}
        print(link_dict)
        all_movie_info.append(link_dict)
    except:
        print(movie + " not found")
```

## Part 3 - Trying with every single movie

```python
import re
import requests
from bs4 import BeautifulSoup

head={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
import time
all_movie_info = []
movies = df.movie.unique()
for movie in movies:
    clean_movie = re.sub(r"[^\w\d\s]","",movie)
    url ="https://www.boxofficemojo.com/search/?q=" + clean_movie.replace(" ","+") 
    print(url)
    raw_html = requests.get(url,headers=head).content
    time.sleep(2)
    soup_doc = BeautifulSoup(raw_html, "html.parser")
    try:
        link = soup_doc.find(class_="a-color-secondary").previous_sibling['href']
        link_dict = {"movie": movie, "link": link}
        print(link_dict)
        all_movie_info.append(link_dict)
    except:
        print(movie + " not found")
```

```python
len(all_movie_info)
```

```python
df_movie_links = pd.DataFrame(all_movie_info)
df_movie_links.to_csv(r'backup_BBC_movie_links.csv', index = False)
```

## Part 4 - Loop with how much money they made

```python
head={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
import time

for movie in all_movie_info[:5]:
    url ="https://www.boxofficemojo.com" + movie['link']
    print(url)
    raw_html = requests.get(url,headers=head).content
    time.sleep(2)
    soup_doc = BeautifulSoup(raw_html, "html.parser")
    try:
        money = soup_doc.h2.find_next_siblings()
        worldwide = money[2].find_all('span')[1].text
        movie['money'] = worldwide
        print(worldwide)
    except:
        print(movie + " not found")
```

Now with all the movies
