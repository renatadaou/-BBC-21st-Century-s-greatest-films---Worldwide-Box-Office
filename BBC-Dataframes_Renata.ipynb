```python
import json
from pandas import json_normalize
with open('world.geojson') as json_data:
    geometry_data = json.load(json_data)
df_geo = pd.DataFrame.from_dict(json_normalize(geometry_data['features']), orient='columns')
df_geo
```

```python
df_geo['properties.NAME_LONG'].replace({'United States of America': 'US',
                                         'United Kingdom': 'UK',
                                         'United Arab Emirates': 'UAE'}, inplace=True)
df_geo.tail(50)
```

```python
final_merged_df = pd.merge(output, df_geo, left_on='critic_country', right_on='properties.NAME_LONG', how='left')
final_merged_df
```

```python
ok_json = json.loads(final_merged_df.to_json(orient='records'))
ok_json
```

```python
def process_to_geojson(file):
    geo_data = {"type": "FeatureCollection", "features":[]}
    for row in file:
        this_dict = {"type": "Feature", "properties":{}, "geometry": {}}
        for key, value in row.items():
            key_names = key.split('.')
            if key_names[0] == 'geometry':
                this_dict['geometry'][key_names[1]] = value
            if str(key_names[0]) == 'properties':
                this_dict['properties'][key_names[1]] = value
        geo_data['features'].append(this_dict)
    return geo_data
```

```python
geo_format = process_to_geojson(ok_json)
geo_format
```
