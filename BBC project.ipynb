### Part 1 - BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup
```

```python
my_url = "https://www.bbc.com/culture/article/20160819-the-21st-centurys-100-greatest-films-who-voted"
raw_html = requests.get(my_url).content
```

```python
soup_doc = BeautifulSoup(raw_html, "html.parser")
print(soup_doc.prettify())
```

```python
sections = soup_doc.find_all('section', class_='sc-4e574cd-0 bhtqwj')

critics_list = []

for section in sections:
    critic_name = section.find('b', attrs={'class': 'sc-7dcfb11b-0 kVRnKf'})
    movies = section.find('p', attrs={'class': 'sc-eb7bd5f6-0 gDEGTg'})
    critic_dict = {'critic_name': critic_name, 'movies': movies}

    critics_list.append(critic_dict)
    
for critic in critics_list:
    print(critic)
```

```python
critics = soup_doc.find_all('b')
for critic in critics:
    print(critic.text)
```

```python
movies = soup_doc.find_all('p')
for movie in movies:
    print(movie.text)
```

### Part 2 - separate the categories with regex

```python
import re
```

#### First, just the movies stuff

```python
movies_text = "\n".join(str(item) for item in movies)

mov = re.findall(r'\d+\. (.+?) \(', movies_text)

print(mov)
```

```python
for line in mov:
    print(line)
```

```python
movies_text = "\n".join(str(item) for item in movies)

director_names = re.findall(r'\d+\..+?\((.+?), \d{4}\)', movies_text)
print(director_names)
```

```python
for director in director_names:
    print(director)
```

```python
movies_text = "\n".join(str(item) for item in movies)

years = re.findall(r'\d+\..+?\(.+?, (\d{4})\)', movies_text)
print(years)
```

```python
for year in years:
    print(year)
```

#### second, just the critics stuff

```python
critics_text = "\n".join(str(item) for item in critics)

names = re.findall(r'^([^\â€“]+) â€“', critics_text, re.MULTILINE)

for name in names:
    print(name)
```

```python
critics_text = "\n".join(str(item) for item in critics)

countries = re.findall(r'\((\w+)\)', critics_text)

for country in countries:
    print(country)
```

```python
critics_text = "\n".join(str(item) for item in critics)

critic_types = re.findall(r'â€“ (.+?) \(', critics_text)

for critic_type in critic_types:
    print(critic_type.strip())
```
